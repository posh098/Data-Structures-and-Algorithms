# Binary Search Tree (All operations)

```python
%%writefile bst.c

#include <stdio.h>
#include <stdlib.h>

// Define structure for a tree node
struct node {
    int data;
    struct node *left, *right;
};

// Create a new node
struct node* createNode(int data) {
    struct node* newNode = (struct node*)malloc(sizeof(struct node));
    newNode->data = data;
    newNode->left = newNode->right = NULL;
    return newNode;
}

// Insert a new node into BST
struct node* insert(struct node* root, int data) {
    if (root == NULL)
        return createNode(data);

    if (data < root->data)
        root->left = insert(root->left, data);
    else if (data > root->data)
        root->right = insert(root->right, data);

    return root;
}

// Find minimum value node (used in deletion)
struct node* findMin(struct node* root) {
    while (root->left != NULL)
        root = root->left;
    return root;
}

// Delete a node
struct node* deleteNode(struct node* root, int data) {
    if (root == NULL)
        return root;

    if (data < root->data)
        root->left = deleteNode(root->left, data);
    else if (data > root->data)
        root->right = deleteNode(root->right, data);
    else {
        // Node found
        if (root->left == NULL) {
            struct node* temp = root->right;
            free(root);
            return temp;
        } else if (root->right == NULL) {
            struct node* temp = root->left;
            free(root);
            return temp;
        }

        struct node* temp = findMin(root->right);
        root->data = temp->data;
        root->right = deleteNode(root->right, temp->data);
    }
    return root;
}

// Search a node
struct node* search(struct node* root, int key) {
    if (root == NULL || root->data == key)
        return root;

    if (key < root->data)
        return search(root->left, key);

    return search(root->right, key);
}

// Inorder traversal (Left, Root, Right)
void inorder(struct node* root) {
    if (root != NULL) {
        inorder(root->left);
        printf("%d ", root->data);
        inorder(root->right);
    }
}

// Preorder traversal (Root, Left, Right)
void preorder(struct node* root) {
    if (root != NULL) {
        printf("%d ", root->data);
        preorder(root->left);
        preorder(root->right);
    }
}

// Postorder traversal (Left, Right, Root)
void postorder(struct node* root) {
    if (root != NULL) {
        postorder(root->left);
        postorder(root->right);
        printf("%d ", root->data);
    }
}

// Function to print the tree structure (sideways view)
void displayTree(struct node* root, int space) {
    if (root == NULL)
        return;

    // Increase distance between levels
    space += 5;

    // Print right child first
    displayTree(root->right, space);

    // Print current node after space
    printf("\n");
    for (int i = 5; i < space; i++)
        printf(" ");
    printf("%d\n", root->data);

    // Print left child
    displayTree(root->left, space);
}

// Main function
int main() {
    struct node* root = NULL;
    int choice, val;

    while (1) {
        printf("\n\n--- Binary Search Tree Menu ---\n");
        printf("1. Insert\n2. Delete\n3. Search\n4. Inorder\n5. Preorder\n6. Postorder\n7. Exit\n");
        printf("Enter your choice: ");
        scanf("%d", &choice);

        switch (choice) {
        case 1:
            printf("Enter value to insert: ");
            scanf("%d", &val);
            root = insert(root, val);
            printf("\nTree after insertion:\n");
            displayTree(root, 0);
            break;

        case 2:
            printf("Enter value to delete: ");
            scanf("%d", &val);
            root = deleteNode(root, val);
            printf("\nTree after deletion:\n");
            displayTree(root, 0);
            break;

        case 3:
            printf("Enter value to search: ");
            scanf("%d", &val);
            if (search(root, val))
                printf("%d found in the tree.\n", val);
            else
                printf("%d not found.\n", val);
            printf("\nCurrent Tree:\n");
            displayTree(root, 0);
            break;

        case 4:
            printf("Inorder Traversal: ");
            inorder(root);
            printf("\n\nCurrent Tree:\n");
            displayTree(root, 0);
            break;

        case 5:
            printf("Preorder Traversal: ");
            preorder(root);
            printf("\n\nCurrent Tree:\n");
            displayTree(root, 0);
            break;

        case 6:
            printf("Postorder Traversal: ");
            postorder(root);
            printf("\n\nCurrent Tree:\n");
            displayTree(root, 0);
            break;

        case 7:
            exit(0);

        default:
            printf("Invalid choice!\n");
        }
    }

    return 0;
}
```
